```python
import pandas as pd
import joblib
from sklearn.metrics import classification_report

# Load model
model = joblib.load("xgboost_binary_model.pkl")

# Load the real-time simulation dataset
df_sample = pd.read_csv("real_time_data_binary.csv")

# Separate features and labels
X_real_time = df_sample.drop(columns=['label'])
y_true = df_sample['label']

# Simulate real-time prediction
y_pred = []
for i, row in X_real_time.iterrows():
    pred = model.predict([row])[0]
    y_pred.append(pred)
    print(f"Message {i+1}: {'Anomaly' if pred == 1 else 'Normal'} | Actual: {'Anomaly' if y_true.iloc[i] == 1 else 'Normal'}")

# Optional: Evaluate
print("\n--- Real-Time Prediction Evaluation ---")
print(classification_report(y_true, y_pred))
```

```python
import datetime

df_sample["predicted_label"] = y_pred
df_sample["timestamp"] = [datetime.datetime.now().isoformat()] * len(df_sample)
df_sample.to_csv("real_time_predictions_log.csv", index=False)
```
