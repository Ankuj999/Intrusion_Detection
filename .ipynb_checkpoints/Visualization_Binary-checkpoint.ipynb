```python
import pandas as pd
import joblib
import time
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.metrics import classification_report, confusion_matrix
from IPython.display import clear_output, display
```

```python
model = joblib.load("D:\\SI Rourkela\\Jup\\xgboost_binary_model.pkl")

# Load your sample CSV for real-time simulation
df = pd.read_csv("D:\\SI Rourkela\\Jup\\real_time_data_binary.csv")

# Separate features and labels
X = df.drop(columns=['label'])
y_true = df['label'].values
```

```python
y_pred = []
confidence_scores = []
anomaly_count = 0

print("--- Starting Real-Time IoMT Anomaly Detection Simulation ---\n")

for i in range(len(X)):
    row = X.iloc[i:i+1]
    pred = model.predict(row)[0]
    proba = model.predict_proba(row)[0][1]  # Confidence for class 1 (Anomaly)
    
    y_pred.append(pred)
    confidence_scores.append(proba)
    
    if pred == 1:
        anomaly_count += 1

    # Simulate live dashboard (replace previous output)
    clear_output(wait=True)
    print(f"üîç Message {i+1}/{len(X)}")
    print(f"‚ñ∂Ô∏è  Prediction: {'Anomaly' if pred == 1 else 'Normal'}")
    print(f"‚úÖ  Actual: {'Anomaly' if y_true[i] == 1 else 'Normal'}")
    print(f"üìä  Confidence: {round(proba, 4)}")
    print(f"üö®  Total Anomalies Detected: {anomaly_count}")
    
    time.sleep(0.3)  # Simulate delay
```

```python
print("\n--- Final Evaluation ---")
print(classification_report(y_true, y_pred, target_names=["Normal", "Anomaly"]))
```

```python
cm = confusion_matrix(y_true, y_pred)

plt.figure(figsize=(6, 4))
sns.heatmap(cm, annot=True, fmt='d', cmap='YlGnBu', 
            xticklabels=["Normal", "Anomaly"], yticklabels=["Normal", "Anomaly"])
plt.title("Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()
```

```python
plt.figure(figsize=(10, 4))
plt.plot(confidence_scores, label="Anomaly Confidence Score", color='darkorange')
plt.axhline(0.5, color='red', linestyle='--', label="Threshold = 0.5")
plt.title("Model Confidence Over Time")
plt.xlabel("Message Index")
plt.ylabel("Confidence Score")
plt.legend()
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
df_results = df.copy()
df_results['predicted'] = y_pred
df_results['confidence'] = confidence_scores
df_results.to_csv("real_time_prediction_results.csv", index=False)
```
